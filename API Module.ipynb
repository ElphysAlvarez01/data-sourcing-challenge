### Merge and Clean the Data for Export

```python
final_merged_df = pd.merge(tmdb_df,reviews_df,on="title")
final_merged_df.head(5)
```

```python

columns_to_fix = ["genre", "spoken_languages", "production_countries"]
characters_remove = ["[", "]", "'"]
    
# Loop through the list of columns to fix
for column in columns_to_fix:
    # Convert the column to type 'str'
    final_merged_df[column] = final_merged_df[column].astype("str")

    # Loop through characters to remove
    for character in characters_remove:
        final_merged_df[column] = final_merged_df[column].str.replace(character, "", regex=False)

# Display the fixed DataFrame
final_merged_df.head()
```

```python
# Drop "byline.person" column
final_merged_df.drop("byline.person", axis=1,inplace=True)
```

```python
# Delete duplicate rows and reset indcle
cleaned_df = final_merged_df.drop_duplicates().reset_index(drop=True)
```
